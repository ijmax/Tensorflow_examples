# Generate artificial dataset for classification

**sklearn.datasets** package has several functions to load or generate datasets for test machine learning algorithms. Some simple functions have been demostrated here.

More details: http://scikit-learn.org/stable/datasets/index.html

````python
%matplotlib inline
import numpy as np
import sklearn.datasets as skd
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
````

There is a function in **scikit learn library** to generate artificial datasets for the machine learning.
```python
sklearn.datasets.make_classification(n_samples=100, n_features=20, n_informative=2, n_redundant=2, n_repeated=0, n_classes=2, n_clusters_per_class=2, weights=None, flip_y=0.01, class_sep=1.0, hypercube=True, shift=0.0, scale=1.0, shuffle=True, random_state=None)[source]
```
Reference: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html

````python
#x,y = skd.make_classification()
x,y = skd.make_classification(n_features=5, n_informative=2, n_redundant=0, n_repeated=0)
````

Color map can be changed using **cmap** attribute. In order to set color map **pylab** package should be imported. Names of the color maps can be found <a href="https://matplotlib.org/examples/color/colormaps_reference.html">here</a>

````python
plt.scatter(x[:,0], x[:,1], c=y, cmap=pl.cm.bwr)
plt.show()
````

## Plotting scatter matrix using pandas
**pandas** library provides a function to generate scatter matrix which help to plot all the possible binary plots for the multidimentional dataset.
```python
pandas.plotting.scatter_matrix(frame, alpha=0.5, figsize=None, ax=None, grid=False, diagonal='hist', marker='.', density_kwds=None, hist_kwds=None, range_padding=0.05, **kwds)
```

Reference: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.plotting.scatter_matrix.html

````python
#data = np.hstack((x,y.T))
df = pd.DataFrame(x)
pd.plotting.scatter_matrix(df, alpha=0.7, figsize=(8, 8), diagonal='kde', s=30, c=y, marker = 'D')
plt.show()
````

## Generate blob dataset
```python
sklearn.datasets.make_blobs(n_samples=100, n_features=2, centers=3, cluster_std=1.0, center_box=(-10.0, 10.0), shuffle=True, random_state=None)
```

Reference: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

````python
x, y = skd.make_blobs(n_samples=100, centers=3, n_features=2, cluster_std=2.0)
colors = {0:'red', 1:'blue', 2:'green'}
plt.scatter(x[:,0], x[:,1], c=y, cmap=pl.cm.rainbow)
plt.show()
````

## Generate gaussian quantiles
```python
sklearn.datasets.make_gaussian_quantiles(mean=None, cov=1.0, n_samples=100, n_features=2, n_classes=3, shuffle=True, random_state=None)
```

Reference: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_gaussian_quantiles.html#sklearn.datasets.make_gaussian_quantiles

````python
x, y = skd.make_gaussian_quantiles()
plt.scatter(x[:,0], x[:,1], c=y, cmap=pl.cm.brg)
plt.show()
````

## Circular distribution

```python
sklearn.datasets.make_circles(n_samples=100, shuffle=True, noise=None, random_state=None, factor=0.8)
```

Reference: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html#sklearn.datasets.make_circles

````python
x, y = skd.make_circles(noise=0.08, factor=0.4)
plt.scatter(x[:,0], x[:,1], c=y, cmap=pl.cm.bwr)
plt.show()
````
